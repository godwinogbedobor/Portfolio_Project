```python

import pandas as pd
```

```python
df = pd.read_excel(r"C:\Users\Godwin Ogbedobor\Downloads\Customer Call List.xlsx")
df
```

```python
#Remove Duplicates
df = df.drop_duplicates()
```

```python
#Delete Columns that are not needed/useful
df=df.drop(columns = ['Not_Useful_Column'])
df
```

```python
#Split address into street, state and zipcode
df[['Street','State','Zipcode']] = df['Address'].str.split(',',expand=True)
df
```

```python
#Remove Address Column
df = df.drop(columns = ['Address'])
df
```

```python
#Remove unwanted characters from the Last_Name Column
df['Last_Name'] = df['Last_Name'].str.strip('123.../_')
df
```

```python

#Standardizing columns with yes and no to only show 'Y' or 'N'
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes','Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No','N')


df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes','Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No','N')
```

```python
df
```

```python
#Standardizing the Phone_number column using regular expression
df['Phone_Number'] = df['Phone_Number'].replace('[^a-zA-Z0-9]','',regex=True)

#Formating the number into string 
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x))
#Standardized Phone numbers
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3]+'-' + x[3:6]+'-' + x[6:10])

df['Phone_Number'] = df['Phone_Number'].replace('nan--','').replace('Na--','')

df
```

```python
#Clearing cells that contain no data
df=df.replace('N/a','')
df = df.fillna('')
df
```

```python
#Returning only entries where Do_Not_Contact is 'N'
for x in df.index:
    if df.loc[x, 'Do_Not_Contact'] == 'Y':
        df.drop(x, inplace=True)
df
```

```python
#Dropping null values int the Phone_Number 
for x in df.index:
    if df.loc[x, 'Phone_Number'] == '':
        df.drop(x, inplace=True)
df
```

```python
#Resetting the index
df = df.reset_index(drop=True)
df
```
